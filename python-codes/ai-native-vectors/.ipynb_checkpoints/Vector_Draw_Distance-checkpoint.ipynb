# Vector Drawing and Distance Calculations

## Imports
Below are the required libraries for the code to run.

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Vector Diagram
It will draw the vector diagram of colors in a 3d vector space.

```python
x1 = 40
y1 = 120
z1 = 60
x2 = 60
y2 = 50
z2 = 90

colour3 = np.array([x1+50,y2+10,z2+20])

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(x1,y1,z1,zdir='z',label='Colour1')
ax.scatter(x2,y2,z2,zdir='z',label='Colour2')
ax.scatter(colour3[0],colour3[1],colour3[2],zdir='z',label='Colour3')
ax.scatter(0,0,0,zdir='z',label='Origin',color='black')

ax.quiver(0,0,0,x1,y1,z1,color='black',arrow_length_ratio=0.1)
ax.quiver(0,0,0,x2,y2,z2,color='black',arrow_length_ratio=0.1)
ax.quiver(0,0,0,colour3[0],colour3[1],colour3[2],color='black',arrow_length_ratio=0.1)

ax.legend()

ax.set_xlim(0,150)
ax.set_ylim(0,150)
ax.set_zlim(0,150)

ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('z-axis')

ax.view_init(elev=10,azim=-35,roll=0)
plt.show()
```

## Vector Distance Calculations
Below code will calculate the Euclidean and Manhattan Distances, Dot Product, and Cosine Similarity

```python
# Euclidean distance
E1 = np.array([(x2-x1)**2,(y2-y1)**2,(z2-z1)**2])
eud_dist = np.sqrt(E1.sum())
# eud_dist = np.linalg.norm(([x1,y1,z1] - [x2,y2,z2]),ord=2)
# eud_dis = sqrt((x2-x1)**2 + (y2-y1)**2 + (z2-z1)**2)
print(eud_dist)

# Manhattan distance
M1 = np.array([np.abs(x1-x2),np.abs(y1-y2),np.abs(z1-z2)])
manhattan_dist = M1.sum()
# manhattan_dist = np.linalg.norm(([x1,y1,z1] - [x2,y2,z2]),ord=1)
# manhattan_dis = np.abs(x1-x2) + np.abs(y1-y2) + np.abs(z1-z2)
print(manhattan_dist)

# Cosine similarity and cosine dissimilarity/distance
cosine_sim = np.dot([x1,y1,z1],[x2,y2,z2])/(np.linalg.norm([x1,y1,z1],ord=2)*np.linalg.norm([x2,y2,z2])) # default: L2(Euclidean), ord=2
# cosine_sim = (x1*x2 + y1*y2 + z1*z2)/(sqrt(x1**2+y1**2+z1**2)*sqrt(x2**2+y2**2+z2**2))
print(cosine_sim)
cosine_dis = 1 - cosine_sim
print(cosine_dis)

# Dot product
dot_product = np.dot([x1,y1,z1],[x2,y2,z2])
# dot_product = x1*x2 + y1*y2 + z1*z2
print(dot_product)
```

## Find the closest vector from Color3
We'll find the closest color from color3

```python

# Cosine similarity from colour1 to colour3
cosine_sim_1_3 = np.dot(colour3,np.array([x1,y1,z1]))/(np.linalg.norm(colour3)*np.linalg.norm([x1,y1,z1]))

# Cosine similarity from colour2 to colour3
cosine_sim_2_3 = np.dot(colour3,np.array([x2,y2,z2]))/(np.linalg.norm(colour3)*np.linalg.norm([x2,y2,z2]))

print('Color1 is closer to Color3') if cosine_sim_2_3 > cosine_sim_1_3 else print('Color2 is closer to Color3')
```
